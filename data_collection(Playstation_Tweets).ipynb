```python

import twint #importing twint package so we can webscrape tweets from twitter

# Configure
c = twint.Config()
c.Search = "Sony Playstation" #We are webscraping tweets from sony playstation
c.Lang = "english"
c.Limit = 1000000
c.Output = "./Sony_Playstation_tweets.csv" #Outputting tweets into a csv file
c.Store_csv = True

# Run
twint.run.Search(c) #Printing out the tweets we got from the webscrape
```
